# Dados de lentidão (2001-2020)


## Abrindo e processando os dados

```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
%matplotlib inline  
from tqdm.notebook import tqdm
```

```python
directory = '../dados/lentidao/baixados/'
#List that will keep the slowness DataFrames
dataframes_list = []
#Iterating over each Slowness file of each year
for filename in os.listdir(directory):
    #Opening the slowness DataFrame
    df = pd.read_csv(directory+filename,sep=';',encoding='latin-1')
    #Verifying if the DataFrame doesn't have the "region" column.
    if len(df.columns)==6:
        #If it doesn't, add this columns (with null values)
        df['nome_regiao'] = np.nan
    df.columns = ['data','corredor','sentido','expressa','trecho','lentidao','regiao']
    #Adding the DataFrame to the list
    dataframes_list.append(df)
#Concating the DataFrames of all years
lentidao_df = pd.concat(dataframes_list,axis = 0, ignore_index=True)
#Setting date column as datetime
lentidao_df['data'] = pd.to_datetime(lentidao_df['data'], dayfirst=True)
lentidao_df['trecho'] = lentidao_df['trecho'].str.strip()
lentidao_df.head(3)
```

## Plotando a lentidão total por ano

```python
fig,ax = plt.subplots(figsize = (16,6))
total_slowness_by_year = lentidao_df.groupby(lentidao_df['data'].dt.year).agg('sum')
ax.plot(total_slowness_by_year)
#ax.set_title('Total Slowness by year', fontsize = 16)
ax.set_xticks(lentidao_df['data'].dt.year.unique())
ax.set_xticklabels(labels =lentidao_df['data'].dt.year.unique())
ax.set_ylabel('Slowness', fontsize=14)
plt.savefig(r'D:\Trabalho\faculdade\TCC\outputs\graficos\analise_exploratoria\slowness_time_series_2001_2020.png')
plt.show()
```

#### Função que arredonda um datetime para a próxima meia hora mais próxima

Isso é feito para compatibilizar os horários do dado de alagamentos com os de lentidão, que são medidos de meia em meia hora

### Relação entre alagamentos (1h antes da lentidão ser medida) e lentidao a cada meia hora

```python
from datetime import datetime, timedelta

def ceil_dt(dt, minutes):
    delta = timedelta(minutes=minutes)
    return dt + (datetime.min - dt) % delta
```

```python
data_apos_1h_list = []
for data in alagamentos_2019['data_inicio']:
    data_apos_1h = (data+ pd.to_timedelta(1, unit='h'))
    data_apos_1h = ceil_dt(datetime(data_apos_1h.year, data_apos_1h.month, data_apos_1h.day, data_apos_1h.hour, data_apos_1h.minute), 30)
    data_apos_1h_list.append(data_apos_1h.strftime("%m/%d/%Y %H:%M:%S"))
```

```python
alagamentos_2019['data_apos_1h'] = pd.to_datetime(data_apos_1h_list)
```

Adicionando colunas de turno do dia, dia da semana ou fim de semana, e estação do ano

```python
merge['dia'] = merge['data_medicao'].dt.weekday.apply(lambda x: 'dia de semana' if x<4 else 'fim de semana')


def get_daytime(x):
    if x in range(0,7):
        return 'madrugada'
    elif x in range(7,13):
        return 'manhã'
    elif x in range(13,19):
        return 'tarde'
    elif x in range(19,24):
        return 'noite'
        
merge['periodo do dia'] = merge['data_medicao'].dt.hour.apply(lambda x: get_daytime(x))

def get_season(x):
    
    spring = range(80, 172)
    summer = range(172, 264)
    fall = range(264, 355)
    # winter = everything else

    if x in spring:
        return'primavera'
    elif x in summer:
        return'verão'
    elif x in fall:
        return 'outono'
    else:
        return 'inverno'
    
    
merge['estação'] = merge['data_medicao'].dt.dayofyear.apply(lambda x: get_season(x))
merge.head(3)
```

### Treinando o modelo

Importando as bibliotecas

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

Separando em treino e teste

```python
X = pd.concat([pd.get_dummies(merge[['dia','periodo do dia', 'estação']]), merge[['alag_ate_3h_antes', 'acidente_ate_3h_antes','tamanho_via','chuva_diaria']]], axis=1)
y = merge['lentidao']
x_train, x_test, y_train, y_test = train_test_split(X,y)
x_train.head(3)
```

Criando o modelo

```python
#model = DecisionTreeRegressor()
model = RandomForestRegressor()
#model = LinearRegression()
model.fit(x_train,y_train)
```

Métricas dos dados de teino


```python
y_pred = model.predict(x_train)
fig, ax = plt.subplots(figsize = (10,6))
r2_treino = r2_score(y_train, y_pred)
print(f'R2 de treino: {round(r2_treino,2)}')
ax.scatter(y_train, y_pred)
plt.show()
```

Métricas dos dados de teste

```python
y_pred = model.predict(x_test)
fig, ax = plt.subplots(figsize = (10,6))
r2_teste = r2_score(y_test, y_pred)
print(f'R2 de teste: {round(r2_teste,2)}')
ax.scatter(y_test, y_pred)
plt.show()
```
